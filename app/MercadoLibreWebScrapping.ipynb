```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
import re
```

```python
# Get data from website
url = "https://listado.mercadolibre.com.mx/deportes-fitness/tenis-paddle-squash/equipamiento/bandas-cabeza/nuevo/bandas-deportivas-para-cabello"
```

```python
page_number = 1
products = []

while True:
    # SIMULA LA NAVEGACIÓN
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
    }
    
    # Agrega el número de página a la URL
    response = requests.get(f'{url}_Desde_{(page_number - 1) * 10}_NoIndex_True', headers=headers)
    # response = requests.get(f'{url}_Desde_100_NoIndex_True', headers=headers)
    
    if response.status_code != 200:
        print('Error', page_number)
        break
    
    html = BeautifulSoup(response.text, 'html.parser')

    allData = html.find_all('div', class_="poly-card__content")
    
    if not allData:
        print(f"Total {page_number}")
        break
    
    for data in allData:
        
        description = data.find('h2', class_="poly-box poly-component__title").text
        s = data.find('s', class_="andes-money-amount andes-money-amount--previous andes-money-amount--cents-dot")
        div = data.find('div', class_ = "poly-price__current")
        discountPrice = div.find('span', class_ = "andes-money-amount andes-money-amount--cents-superscript").text.replace("$","").replace(",",".")
        normalPrice = s.text.replace("$","").replace(",",".") if s else None
        a = data.find('a')
        link = a['href'] if a else None
        
        products.append({ "Description" : description ,
                          "Normal Price" :  normalPrice if normalPrice else discountPrice,
                          "Discount Price" : discountPrice if normalPrice is not None else normalPrice,
                          "URL" : a['href'] if a else None
                        })
    page_number += 1
    time.sleep(2)
        
        
#print(products)        
df = pd.DataFrame(products)
# df.to_csv('products.csv')
```

```python
from functools import reduce

products = []
start = 0
product = "cerave"
szPerPage = 54
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
}

# while page<end:
# product = "crema"
page = 1
replaces = [{"pattern":"\\$","target":""}, {"pattern":",","target":""},{"pattern":"\n","target":""},{"pattern":" ","target":""},{"pattern":"\\(Oferta\\)","target":""}]

def replace_values(string:str, values:dict)->str:
    pattern = values["pattern"]
    target = values["target"]
    return re.sub(pattern, target, string).strip()
    

while True:
    # URL = f"https://www.yza.mx/on/demandware.store/Sites-YzaMexico-Site/es_MX/Search-UpdateGrid?&start={(start*szPerPage)}&sz={szPerPage}"
    URL = f"https://www.yza.mx/busqueda?q={product}&search-button=&lang=es_M&start={((start*szPerPage))}&sz=54&maxsize=54"
    print(URL)
    response = requests.get(URL, headers=headers)
    html = BeautifulSoup(response.text, 'html.parser')
    items = html.find_all('div', class_="product product-wrapper")
    if not items:
        break
    
    for item in items:
    
        description = item.find('div', class_="pdp-link").text
        # .replace("\n\n            \n                ","").replace("\n            \n        \n","").replace(" ","", 1)
        description = re.sub(r'\s+', ' ', description).strip()
        
        # print(description.encode('unicode_escape').decode())
        # break
        # link = item.find('a')['href']
        # # oldPriceItem = item.find('span', class_="old-price")
        # oldPriceValue = item.find('span', class_="price-original").text.replace("$","").replace(",","").replace("\n","").replace(" ","").replace("(Antes)", "") if item.find('span', class_="price-original") else None
        # currentPrice = item.find('span', class_="value").text.replace("$","").replace(",","").replace("\n","").replace(" ","").replace("(Oferta)", "")
        currentPrice = item.find('span', class_="value").text
        currentPrice = reduce(replace_values, replaces, currentPrice)
        products.append({
                        "Description" : description,
                        # "Old Price" : oldPriceValue if oldPriceValue else None,
                        "Current Price" : float(currentPrice) if currentPrice else None,
                        # "Link" : f"https://www.yza.mx{link}"
                        })

#     # break
#             # time.sleep(2)
    # break
  
    start +=1
    
df = pd.DataFrame(products)
df
```
